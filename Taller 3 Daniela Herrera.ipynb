# Descripcion de la base

Una red de adyacencias de palabras de adjetivos y sustantivos comunes en la novela "David Copperfield" de Charles Dickens.

# Que?
contiene la red de adjetivos y sustantivos comunes
adyacencias para la novela "David Copperfield" de Charles Dickens

# Por qué?
Los nodos representan los que ocurren con más frecuencia, Los valores de los nodos son 0 para adjetivos y 1
para sustantivos Los bordes conectan cualquier par de palabras que ocurren en posición adyacente
en el texto del libro.

# Como?

Por medio de una red no dirigida se logra ver las asociacioens entre los adjetivos y sustantivos que ocurren con mayor frecuencia

# Visualizacion

## Canales

**Marcas** Nodos/Aristas.

**Canales** posición -horizontal/Vertical.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import networkx as nx
from pyvis.network import Network
import matplotlib.pyplot as plt
```

```python
g = nx.read_gml('adjnoun.gml')
```

```python
plt.figure(figsize=(60,25))
plt.subplot(121)
nx.draw(g, with_labels=True,node_color='red',width=1,node_size=500,alpha=0.8,linewidths=40)
```
